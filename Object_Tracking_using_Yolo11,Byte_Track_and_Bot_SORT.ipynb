# Detection

```python
from ultralytics import YOLO
model = YOLO("yolo11m.pt")

results = model("/content/p.mp4", save =True)
```

```python
results = model("/content/p.mp4",classes=0, save =True)
```

# Bot Sort Tracker

```python
results = model.track(source="/content/p.mp4", persist= True, save= True)
```

# Byte Track Tracker

```python
results = model.track(source="/content/p.mp4", persist= True, save= True, tracker="bytetrack.yaml")
```
